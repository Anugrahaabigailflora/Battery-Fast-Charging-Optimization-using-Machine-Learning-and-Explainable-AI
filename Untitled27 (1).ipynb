```python
# =====================================================================
# Battery Fast Charging Optimization — Final Explainable AI Version
# =====================================================================
# This version:
# ✅ Adds controlled data noise to prevent overfitting
# ✅ Adds cross-validation for robust evaluation
# ✅ Adds SHAP explainability for feature interpretation
# ✅ Keeps everything realistic and neat
# =====================================================================

import warnings, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder

warnings.filterwarnings("ignore")

# Optional: SHAP for explainability
try:
    import shap
    _HAS_SHAP = True
except:
    _HAS_SHAP = False

# -------------------------
# 0️⃣ CONFIGURATION
# -------------------------
CSV_PATH = '/content/Battery/ev_battery_charging_data.csv'  # Change this path
RANDOM_STATE = 42
FAST_THRESHOLD_MIN = 60  # ≤60 minutes = Fast
np.random.seed(RANDOM_STATE)

# -------------------------
# 1️⃣ LOAD & BASIC CLEANING
# -------------------------
df = pd.read_csv(CSV_PATH)
print("Raw shape:", df.shape)
rename_map = {
    'SOC (%)':'SOC','Voltage (V)':'Voltage','Current (A)':'Current',
    'Battery Temp (°C)':'Battery_Temp','Ambient Temp (°C)':'Ambient_Temp',
    'Charging Duration (min)':'Duration_min','Efficiency (%)':'Efficiency'
}
df.rename(columns={k:v for k,v in rename_map.items() if k in df.columns}, inplace=True)
print("Columns after rename:", list(df.columns))

# Drop rows missing critical info
df.dropna(subset=['SOC','Voltage','Current','Duration_min'], inplace=True)

# Encode categoricals if any
for c in df.select_dtypes(include='object').columns:
    df[c] = LabelEncoder().fit_transform(df[c].astype(str))

# -------------------------
# 2️⃣ FEATURE ENGINEERING
# -------------------------
df['Power_W'] = df['Voltage'] * df['Current']
df['Temp_Diff'] = df['Battery_Temp'] - df['Ambient_Temp']
df['SOC_Remain'] = 100 - df['SOC']
df['Cur_Sq'] = df['Current']**2
df['Power_per_SOC'] = df['Power_W'] / (df['SOC_Remain']+1e-6)
df['Degradation_Risk'] = (df['Current'] * (df['Battery_Temp']-25)) / 100

# Add small controlled noise to reduce overfitting (makes model robust)
df['Duration_min'] += np.random.normal(0, 1.5, len(df))

X = df.select_dtypes(include=[np.number]).drop(columns=['Duration_min'])
y = df['Duration_min']

print(f"Final feature set ({X.shape[1]} features, {len(df)} samples)")

# -------------------------
# 3️⃣ TRAIN / TEST SPLIT + MULTI-MODEL BENCHMARK
# -------------------------
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=RANDOM_STATE)

models = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(n_estimators=250, max_depth=8, random_state=RANDOM_STATE),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=400, learning_rate=0.05, max_depth=6, random_state=RANDOM_STATE)
}

cv_scores = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    cv_scores[name] = np.round(scores.mean(),3)
    print(f"{name} Mean CV R² = {cv_scores[name]}")

best_model_name = max(cv_scores, key=cv_scores.get)
best_model = models[best_model_name]
print(f"\n✅ Best Model Selected: {best_model_name}")

# -------------------------
# 4️⃣ TRAIN FINAL MODEL & EVALUATE
# -------------------------
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

r2 = r2_score(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(f"\nFinal Test R² = {r2:.3f}, RMSE = {rmse:.3f} min")

# Classification metrics (fast vs slow)
y_test_cls = (y_test <= FAST_THRESHOLD_MIN).astype(int)
y_pred_cls = (y_pred <= FAST_THRESHOLD_MIN).astype(int)
acc = accuracy_score(y_test_cls, y_pred_cls)
f1 = f1_score(y_test_cls, y_pred_cls)
print(f"Fast/Slow Accuracy = {acc:.3f}, F1 = {f1:.3f}")
print(classification_report(y_test_cls, y_pred_cls, target_names=['Slow','Fast']))

# -------------------------
# 5️⃣ FEATURE IMPORTANCE
# -------------------------
plt.figure(figsize=(8,5))
if hasattr(best_model, "feature_importances_"):
    imp = pd.Series(best_model.feature_importances_, index=X.columns).sort_values()
    sns.barplot(x=imp, y=imp.index, palette="crest")
    plt.title(f"Feature Importance - {best_model_name}")
    plt.xlabel("Importance")
else:
    coef = pd.Series(best_model.coef_, index=X.columns).sort_values()
    sns.barplot(x=coef, y=coef.index, palette="crest")
    plt.title(f"Feature Coefficients - {best_model_name}")
plt.show()

# -------------------------
# 6️⃣ SHAP EXPLAINABILITY
# -------------------------
if _HAS_SHAP:
    print("Computing SHAP values for Explainable AI...")
    explainer = shap.Explainer(best_model, X_train)
    shap_values = explainer(X_test)
    shap.summary_plot(shap_values, X_test, plot_type="bar")
else:
    print("⚠️ SHAP not installed. Run: pip install shap")

# -------------------------
# 7️⃣ PREDICTED vs ACTUAL
# -------------------------
plt.figure(figsize=(6,5))
plt.scatter(y_test,y_pred,alpha=0.7,edgecolor='k')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--')
plt.title("Predicted vs Actual Charging Duration")
plt.xlabel("Actual Duration (min)")
plt.ylabel("Predicted Duration (min)")
plt.grid(True)
plt.show()

# -------------------------
# 8️⃣ MULTI-OBJECTIVE OPTIMIZATION (Time vs Temp vs Degradation)
# -------------------------
currents = np.linspace(20,160,100)
duration = 1800 / currents
temp = 25 + (0.04*currents) + (0.0015*currents**1.6)
deg = (currents**1.3)/700
safe = temp < 42

score = 0.6*(duration/duration.max()) + 0.25*(temp/temp.max()) + 0.15*(deg/deg.max())
opt_idx = np.argmin(score[safe])
opt_current = currents[safe][opt_idx]

plt.figure(figsize=(8,5))
plt.plot(currents,duration,label="Charging Duration")
plt.plot(currents,temp,label="Battery Temp (°C)")
plt.plot(currents,deg,label="Degradation Index")
plt.axvline(opt_current,color='r',ls='--',label=f"Optimal {opt_current:.1f}A")
plt.axhline(42,color='gray',ls=':',label="Safe Temp = 42°C")
plt.legend();plt.grid(True)
plt.title("Multi-Objective Optimization (Time vs Temp vs Degradation)")
plt.show()

print(f"\n⚡ Recommended Safe Optimal Charging Current ≈ {opt_current:.1f} A (Temp ≈ {temp[safe][opt_idx]:.1f}°C)")

# =====================================================================
# END OF SCRIPT
# =====================================================================
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

st.set_page_config(page_title="Battery Fast Charging Optimization", layout="wide")

st.title("🔋 Battery Fast Charging Optimization (EV)")
st.markdown("This app predicts charging duration and finds the optimal safe current for electric vehicle batteries.")

# ---------------------------
# Sidebar Inputs
# ---------------------------
st.sidebar.header("🔧 Input Parameters")

SOC = st.sidebar.slider("State of Charge (SOC %)", 0, 100, 50)
Voltage = st.sidebar.slider("Voltage (V)", 300, 450, 400)
Current = st.sidebar.slider("Charging Current (A)", 20, 160, 80)
Battery_Temp = st.sidebar.slider("Battery Temp (°C)", 20, 50, 30)
Ambient_Temp = st.sidebar.slider("Ambient Temp (°C)", 15, 40, 25)
Efficiency = st.sidebar.slider("Charging Efficiency (%)", 85, 100, 96)

# ---------------------------
# Mock Model Prediction (replace later with real model if needed)
# ---------------------------
Power_W = Voltage * Current
SOC_Remain = 100 - SOC
Temp_Diff = Battery_Temp - Ambient_Temp

# Simplified duration model (mimicking your ML output)
duration_min = (2000 / max(50, Power_W / 1000)) * (SOC_Remain / 50) * (1 + max(0, Battery_Temp - 30) * 0.02)
duration_min = np.round(duration_min, 1)

# ---------------------------
# Multi-objective Optimization
# ---------------------------
currents = np.linspace(20, 160, 100)
duration_curve = 1800 / currents
temp_curve = 25 + (0.04 * currents) + (0.0015 * currents ** 1.6)
deg_curve = (currents ** 1.3) / 700
safe = temp_curve < 42

score = 0.6 * (duration_curve / duration_curve.max()) + \
        0.25 * (temp_curve / temp_curve.max()) + \
        0.15 * (deg_curve / deg_curve.max())
opt_current = currents[safe][np.argmin(score[safe])]

# ---------------------------
# Output Results
# ---------------------------
st.subheader("⚙️ Prediction Results")
col1, col2, col3 = st.columns(3)
col1.metric("Predicted Duration", f"{duration_min} min")
col2.metric("Power", f"{Power_W:.0f} W")
col3.metric("Optimal Current", f"{opt_current:.1f} A")

# ---------------------------
# Charts
# ---------------------------
st.subheader("📈 Optimization Curves")
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(currents, duration_curve, label="Duration (min)")
ax.plot(currents, temp_curve, label="Battery Temp (°C)")
ax.plot(currents, deg_curve, label="Degradation Index")
ax.axvline(opt_current, color="r", linestyle="--", label=f"Optimal ≈ {opt_current:.1f} A")
ax.axhline(42, color="gray", linestyle=":", label="Safe Temp = 42°C")
ax.legend(); ax.grid(True)
st.pyplot(fig)

# ---------------------------
# Insights Section
# ---------------------------
st.markdown(f"""
### 🔍 Insights
- **Higher Current →** faster charging but increased temperature and degradation.
- **SOC Influence →** charging slows down as SOC increases.
- **Ambient Temperature →** warmer environment leads to higher battery temperature.
- **Efficiency Impact →** lower efficiency means more heat and energy loss.
- **Recommended Current:** **{opt_current:.1f} A** (maintains temperature below 42°C).
""")

st.caption("© 2025 Battery Fast Charging Optimization — Streamlit Edition")
```
